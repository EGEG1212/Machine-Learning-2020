```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
dir(cancer)
```

```python
import pandas as pd

df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df.head()
```

```python
df.columns
```

```python
#정규화
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cancer_scaled = scaler.fit_transform(cancer.data)
scaled_df = pd.DataFrame(data=cancer_scaled, columns=cancer.feature_names)
scaled_df.head()
```

```python
#훈련셋나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    cancer_scaled, cancer.target, test_size=0.2, random_state=2021
)
```

## 서포트벡터머신 모델 생성/학습/예측/평가

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc = SVC()
svc.fit(X_train, y_train)
pred = svc.predict(X_test)
accuracy_score(y_test, pred)
```

```python
svc.get_params()
```

### SVC 모델 0.9912가 나왔다! 더 이상 만지면 나빠질 것 같다;;

```python
#그래도 한번 도전~~!!
from sklearn.model_selection import GridSearchCV

params = {
    'C':[ 0.1, 1, 5, 10, 100]
}

grid_clf = GridSearchCV(svc, param_grid=params, scoring='accuracy', cv=5) 
grid_clf.fit(X_train, y_train)
print(f'최고 평균 정확도: {grid_clf.best_score_:.4f}')
print('최적 파라미터:', grid_clf.best_params_)
```

```python
best_clf = grid_clf.best_estimator_
pred = best_clf.predict(X_test)
accuracy_score(y_test, pred)
```

## 랜덤 포레스트 모델 생성/학습/예측/평가

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
acc = accuracy_score(y_test, pred)
print(f'랜덤 포레스트 정확도: {acc:.4f}')
```

# load_digis

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
import pandas as pd

df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
df.head()
```

```python
#정규화
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
digits_scaled = scaler.fit_transform(digits.data)
scaled_df = pd.DataFrame(data=digits_scaled, columns=digits.feature_names)
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    digits_scaled, digits.target, test_size=0.2, random_state=2021
)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc = SVC()
svc.fit(X_train, y_train)
pred = svc.predict(X_test)
accuracy_score(y_test, pred)
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'C':[ 0.1, 1, 4, 5, 10, 100]
}

grid_clf = GridSearchCV(svc, param_grid=params, scoring='accuracy', cv=5) 
grid_clf.fit(X_train, y_train)
print(f'최고 평균 정확도: {grid_clf.best_score_:.4f}')
print('최적 파라미터:', grid_clf.best_params_)
```

```python
# 한번더 
from sklearn.model_selection import GridSearchCV

params = {
    'C':[ 2, 3, 4, 5]
}

grid_clf = GridSearchCV(svc, param_grid=params, scoring='accuracy', cv=5) 
grid_clf.fit(X_train, y_train)
print(f'최고 평균 정확도: {grid_clf.best_score_:.4f}')
print('최적 파라미터:', grid_clf.best_params_)
# 결과값 정확도 0.9910, 최적 파라미터: {'C': 4}
```

### 최적 파라미터는 'C': 4로구나~~ (정확도 0.9910)

```python
best_clf = grid_clf.best_estimator_
pred = best_clf.predict(X_test)
accuracy_score(y_test, pred)
```
