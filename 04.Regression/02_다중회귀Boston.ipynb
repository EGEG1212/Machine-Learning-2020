## 보스톤 주택 가격 - 다중 회귀

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

### 산점도와 회귀 곡선

```python
fig, asx = plt.subplots(figsize=(16,8), ncols=4, nrows=2)
lm_features = ['RM', 'ZN', 'INDUS', 'NOX', 'AGE', 'PTRATIO','LSTAT', 'RAD']
for i, feature in enumerate(lm_features):
    row, col = int(i/4), i%4
    sns.regplot(x=feature, y='PRICE', data=df, ax=asx[row][col])
    #더 각도가 살아있는것이 영향력이 있다 RM방갯수 정방향 (양의 영향) 
    #LSTAT하위계층는 가격에 (음의 영향)을 끼친다.
```

### Scikit-Learn으로 회귀 모델

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size=0.3, random_state=156
)
```

```python
import numpy as np
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred) 
print(f'MSE: {mse:.2f}, RMSE: {np.sqrt(mse):.2f}')
print(f'R_squared: {r2:.4f}')
```

```python
lr.coef_ #13개 컬럼에 해당하는 계수
```

```python
lr.intercept_
```

### 교차검증 - cross_val_score

```python
from sklearn.model_selection import cross_val_score
lr = LinearRegression()
```

```python
# 5폴드5Fold세트로 MSE를 구함
neg_mean_scores = cross_val_score(lr, boston.data, boston.target, scoring='neg_mean_squared_error', cv=5) #neg_mean_squared_error 때문에 값이 음수로 나온다..
rmse_scores = np.sqrt(-neg_mean_scores)
avg_rmse = np.average(rmse_scores)

print('개별 MSE:', np.round(neg_mean_scores,2))
print('개별 RMSE:', np.round(rmse_scores,2))
print(f'평균 RMSE:, {avg_rmse:.4f}')
```
